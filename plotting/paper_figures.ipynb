### Plot distribution of clone sizes (Sup.Fig.5B)

```python
matplotlib.rcParams['font.family'] = 'Arial'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.labelsize'] = 10
matplotlib.rcParams['xtick.labelsize'] = 10
matplotlib.rcParams['ytick.labelsize'] = 10
matplotlib.rcParams['legend.fontsize'] = 10
```

```python
def plot_clone_size_dist(df_filtered, title, ax):
    bin1 = df_filtered["bin1"]
    bin2 = df_filtered["bin2"]
    bin3 = df_filtered["bin3"]
    bin4 = df_filtered["bin4"]
    bin5 = df_filtered["bin5"]
    bin1_sum = sum(list(bin1))
    bin2_sum = sum(list(bin2))
    bin3_sum = sum(list(bin3))
    bin4_sum = sum(list(bin4))
    bin5_sum = sum(list(bin5))
    total_num = [bin1_sum,bin2_sum,bin3_sum,bin4_sum,bin5_sum]
    cat = ['$< 10^{1}$','$10^{1}$-$10^{2}$','$10^{2}$-$10^{3}$','$10^{3}$-$10^{4}$','$\geq 10^{4}$']
    ax.bar(cat, total_num)
    ax.set_ylim(0,250000) #might need to adjust the ylim max
    ax.set_xlabel("Clone Sizes")
    if "Start" in title:
        ax.set_ylabel("Number of clones")
    ax.set_title(title)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
```

```python
fig, axs = plt.subplots(1, 4, figsize=(16, 4), sharey=True, constrained_layout=True)
plot_clone_size_dist(df_totals[(df_totals["t"]>=60) & (df_totals["t"]<61)], "Start of ART", axs[0])
plot_clone_size_dist(df_totals[(df_totals["t"]>=61) & (df_totals["t"]<62)], "1 month after ART", axs[1])
plot_clone_size_dist(df_totals[(df_totals["t"]>=62) & (df_totals["t"]<63)], "2 months after ART", axs[2])
plot_clone_size_dist(df_totals[(df_totals["t"]>=63) & (df_totals["t"]<64)], "3 months after ART", axs[3])
fig.text(-0.01, 0.95, "B", fontsize=14, fontweight='bold', va='top', ha='left')
# plt.savefig("../figures/for_paper/sup_clone_size_dist.pdf", format='pdf', bbox_inches='tight')
plt.savefig("../figures/for_paper/sup_clone_size_dist.png", dpi = 300, bbox_inches = 'tight')
# plt.savefig("../figures/for_paper/sup_clone_size_dist.eps", format='eps', bbox_inches='tight')
plt.show()
```

### Calculate fold change of large clones (Sup.Fig.6)

```python
large_clones = df_clones[(df_clones["L"]>1000)]

year1 = large_clones[large_clones["t"]==72]
year2 = large_clones[large_clones["t"]==84]
year4 = large_clones[large_clones["t"]==96]

year_1_2 = year1['ID_created'][year1['ID_created'].isin(year2['ID_created'])].unique()
year_1_4 = year1['ID_created'][year1['ID_created'].isin(year4['ID_created'])].unique()

year_1_2_change = {}
year_1_4_change = {}
for ID in list(year_1_2):
    year_1_2_change[ID] = year2.loc[year2['ID_created'] == ID, 'L'].values[0]/year1.loc[year1['ID_created'] == ID, 'L'].values[0]
for ID in list(year_1_4):
    year_1_4_change[ID] = year4.loc[year4['ID_created'] == ID, 'L'].values[0]/year1.loc[year1['ID_created'] == ID, 'L'].values[0]

year_1_2_IDs = list(year_1_2_change.keys())
year_1_4_IDs = list(year_1_4_change.keys())

all_IDs = []
for item in year_1_2_IDs:
    all_IDs.append(item)
for item in year_1_4_IDs:
    if item not in all_IDs:
        all_IDs.append(item)

all_values = {}
for id in all_IDs:
    to_append = []
    if id in list(year_1_2_change.keys()):
        to_append.append(year_1_2_change[id])
    if id in list(year_1_4_change.keys()):
        to_append.append(year_1_4_change[id])
    all_values[id] = to_append

filtered_values = {}
for k,v in all_values.items():
    if len(v)==2:
        filtered_values[k] = v
```

```python
data = []
for clone, (val_2yr, val_4yr) in filtered_values.items():
    data.append({'Clone': clone, 'Year': 2, 'FoldChange': val_2yr})
    data.append({'Clone': clone, 'Year': 4, 'FoldChange': val_4yr})

df = pd.DataFrame(data)

cm2inch = lambda x: x/2.54
ONE_FIVE_COLUMN = cm2inch(8.8)
w     = ONE_FIVE_COLUMN #SLIDE_WIDTH
goldh = w / 3
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Year', y='FoldChange')
plt.yscale('log')
plt.ylabel("Fold change from year 1")
plt.xlabel("Time (Year)")
plt.grid(True, which='both', linestyle='--', alpha=0.3)
plt.tight_layout()
# plt.savefig("../figures/for_paper/sup_fold_change_box_plot.pdf", format='pdf', bbox_inches='tight')
# plt.savefig("../figures/for_paper/sup_fold_change_box_plot.eps", format='eps', bbox_inches='tight')
plt.savefig("../figures/for_paper/sup_fold_change_box_plot.png", dpi = 300)
plt.show()
```

```python
def low_limit(x_dat,y_dat):
    count = 0
    low_thresh = 0
    for item in list(y_dat[0]):
        if item>=200 and list(x_dat[0])[count]>360:
            low_thresh = count
            return low_thresh
        count+=1
    return low_thresh

#writing a method to plot this
lower_hist = []
higher_hist = []
number_of_runs = 0;

#uncomment if early sim:
ORDER_FULLSIM = 10

file_path = "../output/final_runs/early_ATI/"
for item in os.listdir(file_path):
    if "totals" in item:
        df = pd.read_csv(os.path.join(file_path, item), memory_map=True, index_col= False)
        time = df["t"]
        T_cells = df["T"]
        Ltot = df["Ltotal"]
        Atot = df["Atotal"]
        Vtot = df["Vtotal"]
        Events = df["nEvents"]
        react = df["nReact"]
        LxR = df["LXRtotal"]
        t_fullsim         = np.array(df['t'])
        Ltotal_fullsim    = np.array(df['Ltotal'])
        LXRtotal_fullsim  = np.array(df['LXRtotal'])
        Atotal_fullsim    = np.array(df['Atotal'])
        ViralLoad_fullsim = (Atotal_fullsim*(10**(11-ORDER_FULLSIM)))/1680
        
        MONTHSART_FULLSIM = 0.5
        ORDER_FULLSIM = 10
        
        xdat = [(t_fullsim-MONTHSART_FULLSIM)*30]
        ydat = [ViralLoad_fullsim]

        l= low_limit(xdat,ydat)
        low = list(xdat[0])[l]

        if(low-360 >= 0):
            lower_hist.append(low-360)
            number_of_runs+=1

cleaned_list = [x for x in lower_hist if not math.isnan(x)]
lower_hist = cleaned_list

cm2inch = lambda x: x/2.54
ONE_FIVE_COLUMN = cm2inch(8.8)
w     = ONE_FIVE_COLUMN #SLIDE_WIDTH
goldh = w / 3
fig = plt.figure(figsize=(w, goldh))
sns.histplot(lower_hist, bins = 6, kde = True, stat = "percent")
plt.xlabel("Time to rebound (days)")
plt.ylabel("Frequency (% of observations)")
plt.title("Rebound time after 1 year on ART (>200 copies/mL threshold)")
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
fig.text(-0.01, 1.10, "B", fontsize=8, fontweight='bold', va='top', ha='left')
plt.savefig("../figures/for_paper/early_sim_ATI_rebound_time.png", dpi = 300, bbox_inches='tight')
plt.savefig("../figures/for_paper/early_sim_ATI_rebound_time.pdf", bbox_inches='tight')
plt.savefig("../figures/for_paper/early_sim_ATI_rebound_time.eps", format = 'eps', bbox_inches='tight')
plt.show()
```
